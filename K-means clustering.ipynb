```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb

iris_data = sb.load_dataset("iris")
iris_data.head(5)
```

```python
sb.pairplot(iris_data, hue="species", size=2.5)
```

# Cluster on sepal_length vs. petal_length

K-means clustering does a pretty good job of rediscovering the classes in the iris data set.

```python
import random
import math
from collections import defaultdict

K = 3

colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]

all_points = []

for i, row in iris_data.iterrows():
    all_points.append((row.sepal_length, row.petal_length))

def distance(x1, y1, x2, y2):
    return math.sqrt( (x1 - x2)**2 + (y1 - y2)**2 )

def K_means_cluster(all_points, K, visualize=False):
    cluster_centroids = {}
    iteration = 1

    # start with random points as centroid
    for i in range(K):
        random_point = random.randint(0, len(all_points) - 1)
        while random_point in cluster_centroids.values():
            random_point = random.randint(0, len(all_points) - 1)

        cluster_centroids[i] = all_points[random_point]

    while True:
        cluster_points = defaultdict(list)

        # assign all points to nearest cluster centroid
        for point in all_points:
            closest_centroid = 0
            closest_centroid_dist = 1e99

            for centroid_index in cluster_centroids:
                centroid = cluster_centroids[centroid_index]
                centroid_dist = distance(centroid[0], centroid[1], point[0], point[1])
                if centroid_dist < closest_centroid_dist:
                    closest_centroid_dist = centroid_dist
                    closest_centroid = centroid_index

            cluster_points[closest_centroid].append(point)

        # recompute the centroid of each cluster
        new_cluster_centroids = {}
        
        if visualize:
            plt.figure()
            plt.title("Iteration %d" % (iteration))
            plt.xlabel("Sepal length")
            plt.ylabel("Petal length")

        for centroid_index in cluster_points:
            cluster_x = []
            cluster_y = []

            for point in cluster_points[centroid_index]:
                cluster_x.append(point[0])
                cluster_y.append(point[1])

            if visualize:
                plt.scatter(cluster_x, cluster_y, color=colors[centroid_index])

            cluster_x = float(sum(cluster_x)) / len(cluster_points[centroid_index])
            cluster_y = float(sum(cluster_y)) / len(cluster_points[centroid_index])

            new_cluster_centroids[centroid_index] = (cluster_x, cluster_y)

        if visualize:
            plt.show()

        # check if the cluster centroids changed
        no_change = True

        for centroid_index in new_cluster_centroids:
            if new_cluster_centroids[centroid_index] != cluster_centroids[centroid_index]:
                no_change = False
                break

        if no_change:
            break

        cluster_centroids = new_cluster_centroids

        iteration += 1
        
    # return the sum of squared distances between the points and their centroid
    SSE = 0.0
    
    for centroid_index in cluster_points:
        for point in cluster_points[centroid_index]:
            SSE += distance(point[0], point[1],
                            cluster_centroids[centroid_index][0], cluster_centroids[centroid_index][1]) ** 2
            
    return SSE
        
K_means_cluster(all_points, K, visualize=True)
;
```

## Determine optimal K

3 seems to be about the optimal K, which makes sense given there's 3 classes in the iris data set.

```python
K_SSE = []

for K in range(1, 9):
    K_SSE.append(K_means_cluster(all_points, K))
    
plt.figure()
plt.plot(range(1, 9), K_SSE)
plt.title("SSE over K for K-means clustering")
plt.xlabel("K")
plt.ylabel("SSE")
```
